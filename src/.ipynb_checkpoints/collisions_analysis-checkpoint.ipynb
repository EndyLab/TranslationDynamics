```python
def pickleCollisionAggregator(datapath,scaling,norm=False):
    phi_sweep_collisions = list()
    crowders = 427+np.array([0,142,142*2,142*3,142*4,142*5,142*6,142*7,142*8,142*9,142*10,142*11,142*12,142*13])
    ribosomes = list(reversed(range(1,15)))

    phi_sweep = list()
    phi_sweep_crowders = list()
    phi_sweep_ribosomes = list()
    for j in range(len(crowders)):
        for i in range(15-ribosomes[j]):
            expts = [(len(phi_sweep)*4,len(phi_sweep)*4+4)]
            if norm==True:
                phi_sweep_collisions.append(np.divide(collisionPickleAggregator(path+data,expts)[0],ribosomes[j]))
            else:
                phi_sweep_collisions.append(collisionPickleAggregator(datapath,expts)[0])
            phi_sweep.append((crowders[i],ribosomes[j]))
            phi_sweep_crowders.append(crowders[i])
            phi_sweep_ribosomes.append(ribosomes[j])

    phi_sweep_collisions = np.multiply(phi_sweep_collisions,scaling)
    return phi_sweep_collisions
```

```python
datapath = "/Users/Akshay/Dropbox/Life/EndyLab/Research/TranslationDynamics/data/"+"180921_1224"

phi_sweep_collisions = list()
crowders = 427+np.array([0,142,142*2,142*3,142*4,142*5,142*6,142*7,142*8,142*9,142*10,142*11,142*12,142*13])
ribosomes = list(reversed(range(1,15)))
norm=False
scaling=1

phi_sweep_added = list()
phi_sweep_crowders_added = list()
phi_sweep_ribosomes_added = list()

replicates = 4

# Compile analyzed data ----------------------------------------------------------------

for j in range(len(crowders)):
    for i in range(15-ribosomes[j]):
        #Gets set of replicates at specified crowder and ribosome ratio
        expts = [(len(phi_sweep_added)*replicates,len(phi_sweep_added)*replicates+replicates)] 

        #Adds total collisions from pickle that contains already analyzed # collisions, and then records them as added
        phi_sweep_collisions.append(collisionPickleAggregator(datapath,expts)[0])
        phi_sweep_added.append((crowders[i],ribosomes[j]))

        #Creates separate list recording which crowders and ribosomes were added
        phi_sweep_crowders_added.append(crowders[i])
        phi_sweep_ribosomes_added.append(ribosomes[j])

#Scales up number of collisions by a factor if want to report collisions in a longer timescale
phi_sweep_collisions = np.multiply(phi_sweep_collisions,scaling)

# Plotting --------------------------------------------------------------------
fig,ax1 = plt.subplots(1,1,figsize=(10,7))

####Set x (ribosomes), y(crowders), and z(# collisions) as columns in a dataframe
x=np.array(phi_sweep_ribosomes_added)
y=np.array(phi_sweep_crowders_added)
z=[collisions for z_i_arr in phi_sweep_collisions for collisions in z_i_arr]
df = pd.DataFrame.from_dict(np.array([x,y,z]).T)
df.columns = ['num_ribosomes','num_crowders','num_collisions']
df['num_collisions'] = pd.to_numeric(df['num_collisions'])
pivotted_total = df.pivot('num_crowders','num_ribosomes','num_collisions')
sns.heatmap(pivotted_total,cmap='viridis_r',ax=ax1)
#ax1.set_title("# of total collisions in a translation voxel in 1s.")path = "/Users/Akshay/Dropbox/Life/EndyLab/Research/TranslationDynamics/data/"
```

```python
%matplotlib inline
def MSD(path,expt_start,expt_end):
    df_outputs = pd.read_csv(path+"outputMolPosList.txt",sep=" ",header=None) #Add batch processing here potentially
    r2_lists=list()
    for expt_num, row in df_outputs.iterrows():
        if(expt_num>=expt_start and expt_num<expt_end):
            try:
                df = pd.read_csv(path+row[0],delimiter=" ",header=None)
                df.columns=["time","r2","r4"]
                r2_lists.append(np.array(df['r2'].tolist()))
            except:
                print("Missing expt")
    print(len(r2_lists[0]))
    return r2_lists,np.add.reduce(r2_lists)/(len(r2_lists))
   
path='/Users/Akshay/Dropbox/Life/EndyLab/Research/TranslationDynamics/data/'
data = "181031_0028/"

#Ignoring ribosomes, 1e-9
plt.plot(MSD(path+data,0,100)[1],color='blue')
plt.plot(MSD(path+data,100,200)[1],color='red')
plt.plot(MSD(path+data,200,300)[1],color='orange')
plt.plot(MSD(path+data,0,50)[1],color='blue')
plt.plot(MSD(path+data,100,150)[1],color='red')
plt.plot(MSD(path+data,200,250)[1],color='orange')
plt.xlabel("time steps")
plt.ylabel("MSD")
```
