```python
#### Sweep of all other parameters
from analysis_utils import *
%load_ext autoreload
%autoreload
def sweepElongTime(NR_scaledRates = {'k1r':718,'k2f':1475,'k2r_nr':1120,'k3_nr':6,'k4':209}):

    path = "/Users/Akshay/Documents/TranslationDynamics/data/"
    ribs=[4,8,9,9,8,7]

    gr_sim = dict()
    data = "191004_0104/"
    for i in range(1,7):
        sim_i = computeTransportRxnTimes(path+data, 1e15, 1e15, 100*(i-1),100*(i-1)+100,cogtRNANum=i, ribosomeNum = 9,scaling=200,NR_scaling = NR_scaledRates)
        gr_sim[i] = (CellLatencies(sim_i))

    data = "191106_0916/"
    for i in range(7,16):
        sim_i = computeTransportRxnTimes(path+data, 1e15, 1e15, 100*(i-1-6),100*(i-1-6)+100,cogtRNANum=i, ribosomeNum = 9,scaling=200,NR_scaling = NR_scaledRates)
        gr_sim[i] = (CellLatencies(sim_i))

    data = "191106_1041/"
    for i in range(16,25):
        sim_i = computeTransportRxnTimes(path+data, 1e15, 1e15, 100*(i-1-15),100*(i-1-15)+100,cogtRNANum=i, ribosomeNum = 9,scaling=200,NR_scaling =NR_scaledRates)
        gr_sim[i] = (CellLatencies(sim_i))

    data = "191106_1206/"
    for i in range(25,34):
        sim_i = computeTransportRxnTimes(path+data, 1e15, 1e15, 100*(i-1-24),100*(i-1-24)+100,cogtRNANum=i, ribosomeNum = 9,scaling=200,NR_scaling = NR_scaledRates)
        gr_sim[i] = (CellLatencies(sim_i))
    gr_dict = {'gr_1':gr_sim}

    transportRxnResults = transportRxnCalc(gr_dict,gr16_ptRNA,gr16_pCodon)

    search_list = transportRxnResults[0]
    return transportRxnResults[1:]
```
