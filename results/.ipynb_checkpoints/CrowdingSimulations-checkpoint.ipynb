```python
#Algorithm to test distribution of # of incorrect samples for independent 47 sided dice with each side uniform probability (# of incorrect rolls per 1 correct roll)

incorrectRollsArr = list()
for i in range(50000):
    incorrectCounter = 0;
    roll = np.random.randint(0,47)
    while(roll!=9): #42 is the arbitrarily picked "correct" tRNA
        roll = np.random.randint(0,47)
        incorrectCounter+=1;
    incorrectRollsArr.append(incorrectCounter)
```

```python
def countCollisionsWithDelay(df, tau_check, sim_time, ribosome_list, tRNA_list, target_tRNA):
    df_t=df.loc[df.time<sim_time]
    df_t=df_t.loc[df_t['reactantA'].isin(tRNA_list)]
    df_t=df_t.sort_values(['reactantB','time'], ascending=[True,True])
    countCollision_arr=list()
    lenCollision_arr = list()
    for r in ribosome_list:
        df_rt = df_t.loc[df_t['reactantB']==r]
        rxn_count=0;
        rxn_time_arr=list()
        rib_available=True;
        ribavail_time=0
        for i, rxn_i in df_rt.iterrows():
            if rxn_i["time"]>=ribavail_time:
                rib_available=True
            if rib_available is True:
                if rxn_i["reactantA"]==target_tRNA:
                    rxn_time_arr.append(rxn_i["time"])
                    rib_available=False;
                    ribavail_time=rxn_i["time"]+tau_check
                elif rxn_i["reactantA"] in tRNA_list:
                    rib_available=False;
                    ribavail_time=rxn_i["time"]+tau_check
        countCollision_arr.append(rxn_time_arr)   
        lenCollision_arr.append(len(rxn_time_arr))
    return countCollision_arr,lenCollision_arr
```

```python
#Experiment 21
#function of both tau_check as well as mixing time (how long you're willing to wait until every ribosome has got a correct tRNA)
#The average "1st collisions' across all ribosomes is expected to be (1/47)*(simtime/tau_check). But, the question is, how are those 
#'1st collisions' distributed across all ribosomes -- this is based on mixing time. As increase the simtime/mixing time ratio, we expect
#the '1st collisions' across all ribosomes to become closer and closer to mean w/ low STD. NOW, poisson distribution matters, since
#expect the "ideal no transport case" to be poisson((1/47)*(simtime/tau_check))
count_zeros=0
for target in np.arange(8,20):
    collCount21,lenCollision21=countCollisionsWithDelay(expt21_df,tau_check=0.0005,sim_time=0.017,ribosome_list=np.arange(1,8),tRNA_list=np.arange(8,55),target_tRNA=target)
    print(lenCollision21)
    count_zeros+=len(lenCollision21) - np.count_nonzero(lenCollision21)
print(count_zeros)

#tRNA target 8 to 20 90% (want 8 zeros)
#0ms--0.3ms--8 zeros
#0.01ms--1.1ms--10 zeros
#0.1ms--14ms-- 8 zeros

#tRNA target 8 to 20 50% (want 42 zeros)
#0ms -- 0.09ms -- 41 zeros
#0.01ms -- 0.4 ms -- 42 zeros
#0.1ms -- 3ms -- 40 zeros
#0.5ms -- 17ms -- 42 zeros
```
