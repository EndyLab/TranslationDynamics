```python
def avg(col):
    tot=0
    std=0
    for i in range(len(col)):
        tot+=np.average(col[i])
        std+=np.std(col[i])
    return tot/len(col)

def median(col):
    tot=0
    for i in range(len(col)):
       tot+=(np.median(np.array(col[i])))
    return tot/len(col)

def disparity(col):
    low=0
    high=0
    for i in range(len(col)):
        if col[i]<10:
            low+=1
        if col[i]>80:
            high+=1
    return low, high
```
